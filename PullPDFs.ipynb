### This notebook will extract all the pdf links on export.arxiv.org in the Economics section from 2017-2019. Then, it will download each of them as pdfs to a folder called WebscrapedDocs.


There is a pattern in the links.
'https://arxiv.org/pdf/2001.00002.pdf' means the year 2020 (20), in january (01), and the number of the preprint (00002).
Use loop to iterate through each PDF in Econ from the beginning of 2017 to the end of 2019.


```python
from urllib import request
from bs4 import BeautifulSoup
import re
import os
import urllib
```

```python
from datetime import datetime
start = datetime.now()
current_time = start.strftime("%H:%M:%S")
print("Notebook started at", current_time)
```

### Web scrape PDFs

```python
links=[]
year=17#beginning year
while(year<20):#year after end year
    month=1#beginning month
    while(month<13):#for each month
        interval=0
        a=1
        prevcheck=0
        while(a==1):#While on the same month
            site='https://export.arxiv.org/list/econ/'+str(year)+"{:02d}".format(month)+'?skip='+str(interval)+'&show=25'
            response = request.urlopen(site).read()
            soup=BeautifulSoup(response, "html.parser")

            #Iterate through html to find right thing to check if the pages are done
            items=list(soup.children)#.children returns list of html sections
            ins=items[4]
            insideItems=list(ins.children)
            moreIns=insideItems[3]
            moreInsideItems=list(moreIns.children)
            evenMore=moreInsideItems[5]
            evenMoreItems=list(evenMore.children)
            yikes=evenMoreItems[1]
            yikesItems=list(yikes.children)
            almost=yikesItems[11]
            almostItems=list(almost.children)
            nearEnd=almostItems[1]
            nearEndItems=list(nearEnd.children)
            check=int(nearEndItems[0].text[1:-1])

            #if we reached last page
            if(check<=prevcheck):
                a=0
                continue;

            #Add pdfs to links list
            newLinks=soup.find_all('a', href=re.compile(r'(.pdf)'))
            links.extend(newLinks)#Stores links of all pdfs in section
            print(len(links))
            interval+=25
            prevcheck=check
        print(str(month) + " DONE")
        month+=1
    print(str(year)+ "DONE")
    year+=1
```

### Clean PDF names

```python
url_list = []
for i in links:
    if(i['href'].startswith('http')):#if link starts with http
        url_list.append(i['href'])#append link to url_list
    else:
        url_list.append("https://export.arxiv.org" + i['href'] + '.pdf')#otherwise, format it

print(url_list)
```

### Download pdfs to computer

```python
for url in url_list:
    print(url)
    fullfilename = os.path.join('/Users/benjaminpfeffer/Desktop/WebscrapedDocs', url.replace("https://export.arxiv.org/pdf/", ""))#remove start of link so only identifier is left
    print(fullfilename)
    request.urlretrieve(url, fullfilename)#download from the url and save it as the fullfilename
    
#if pdf gives error, find and remove it from url_list, like done above
```
